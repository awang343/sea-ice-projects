```python
all_buoys = get_metadata().index.to_list()

valid_buoys = [buoy for buoy in all_buoys if usable_buoy(buoy)]

interp_lens = pd.DataFrame([{"Sensor ID": buoy, "length": len(buoy_data(buoy, "interp"))} for buoy in valid_buoys])
interp_lens = interp_lens.set_index("Sensor ID")

valid_buoys = get_metadata().merge(interp_lens, how="right", left_index=True, right_index=True)

grouped_stations = valid_buoys[["DN Station ID", "Buoy Type"]].groupby("DN Station ID")


def most_data(df):
    return interp_lens.loc[df.index].idxmax().loc["length"]
def best_buoy(station_df):
    tbuoys = station_df[station_df["Buoy Type"] == "Thermistor"]
    if len(tbuoys) > 0:
        return most_data(tbuoys)
    else:
        return most_data(station_df)

{station[0]: best_buoy(station[1]) for station in grouped_stations}
```
